```python
import pandas as pd
```

```python
Salary = pd.read_csv("Salary_prediction.csv")
```

```python
Salary
```

```python
Independent = Salary[["years_experience"]]
```

```python
Independent
```

```python
print(Salary.columns)
```

```python
Dependent = Salary["salary_per annum "]
```

```python
Dependent
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(Independent,Dependent, test_size=0.2, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor
```

```python
regressor.fit(X_train, y_train)
```

```python
Weight = regressor.coef_
```

```python
Weight
```

```python
Bias = regressor.intercept_
```

```python
Bias
```

```python
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
r_score = r2_score(y_test, y_pred)
```

```python
r_score
```

```python
Salary.columns = Salary.columns.str.strip()

experience = int(float(input("Enter years of experience: ")))
expected_salary = float(input("Enter your expected salary: "))

suggested_companies = Salary[
    (Salary["years_experience"] == experience) &
    (Salary["salary_per annum"] >= expected_salary)
]

if suggested_companies.empty:
    print("No companies found for your expectation 😔")
else:
    print("You can try these companies:")
    print(suggested_companies[["Company_Name","Role","salary_per annum"]])

```
